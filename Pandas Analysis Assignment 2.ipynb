```python
#evaluate average distance from home by jobrole
c1 = df['DistanceFromHome']
c2 = df['JobRole']

pd.crosstab (c1,c2)
```

```python
#evaluate average distance from home by attrition
df[['DistanceFromHome', 'Attrition']].groupby('Attrition').mean()
```

```python
#Employees who live further from the work place have a higher attrition rate
```

```python
#evaluate average monthly income by attrition
df[['MonthlyIncome', 'Attrition']].groupby('Attrition').mean()
```

```python
#Employees with a lower salary have a higher attrition rate
```

```python
#evaluate average monthly income by education
df[['MonthlyIncome', 'Education']].groupby('Education').mean()
```

```python
#find out other factors leading to attrition
```

```python
#find out if rate of business travel affects attrition
c1 = df['BusinessTravel']
c2 = df['Attrition']

pd.crosstab (c1,c2)
```

```python
#find out if employee age affects attrition

df[['Age', 'Attrition']].groupby('Attrition').mean()
```

```python
#Younger employees tend to leave the company as compared to older emploees
```

```python
#find out if years worked at the company affects attrition

df[['YearsAtCompany', 'Attrition']].groupby('Attrition').mean()
```

```python
#Those who have worked for less years tend to leave more than long term employees
```
